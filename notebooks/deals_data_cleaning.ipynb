```python
import ollama
def generate_embedding(text):
    if pd.isna(text) or text.strip() == "":
        return None  # Handle missing values
    
    response = ollama.embeddings(model="nomic-embed-text", prompt=text)
    
    return response["embedding"]  # Extract embedding vector

# 🔹 Apply Embeddings to the DataFrame
deals_df["Title_Embedding"] = deals_df["Title"].apply(generate_embedding)
deals_df["Description_Embedding"] = deals_df["Description"].apply(generate_embedding)
deals_df["Categories_Embeddings"] = deals_df["Categories"].apply(generate_embedding)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# 🔹 Function to Find Similar Deals
def find_similar_deals(content_id, deals_df, top_n=5):
    """
    Find similar deals based on embeddings.
    
    Args:
        content_id (int): The ContentId of the target deal.
        deals_df (pd.DataFrame): DataFrame containing deals and their embeddings.
        top_n (int): Number of similar deals to return.
        
    Returns:
        pd.DataFrame: A DataFrame with the most similar deals.
    """
    # Ensure embeddings exist in DataFrame
    if "Title_Embedding" not in deals_df or "Description_Embedding" not in deals_df or "Categories_Embeddings" not in deals_df:
        raise ValueError("Embeddings not found. Please generate embeddings first.")
    
    # Find the target row
    target_row = deals_df[deals_df["ContentId"] == content_id]
    
    if target_row.empty:
        raise ValueError(f"No deal found with ContentId {content_id}")
    
    # Extract target embeddings (reshape to 2D for cosine_similarity)
    title_embedding = np.array(target_row["Title_Embedding"].values[0]).reshape(1, -1)
    desc_embedding = np.array(target_row["Description_Embedding"].values[0]).reshape(1, -1)
    cat_embedding = np.array(target_row["Categories_Embeddings"].values[0]).reshape(1, -1)

    # Compute similarity for all deals
    title_embeddings = np.vstack(deals_df["Title_Embedding"].values)
    desc_embeddings = np.vstack(deals_df["Description_Embedding"].values)
    cat_embeddings = np.vstack(deals_df["Categories_Embeddings"].values)

    title_similarities = cosine_similarity(title_embedding, title_embeddings)[0]
    desc_similarities = cosine_similarity(desc_embedding, desc_embeddings)[0]
    cat_similarities = cosine_similarity(cat_embedding, cat_embeddings)[0]

    # Compute final similarity score (weighted average)
    final_similarity = (title_similarities + desc_similarities + cat_similarities) / 3

    # Add similarity scores to DataFrame
    deals_df["Similarity"] = final_similarity

    # Sort by similarity (excluding the target deal)
    similar_deals = deals_df[deals_df["ContentId"] != content_id].sort_values(by="Similarity", ascending=False)

    # Return top similar deals
    return similar_deals.head(top_n)
```
