```python
import os
import sys
from pathlib import Path

import pandas as pd
import seaborn as sns

# Setting the working directory to the root of the project
script_dir = Path(os.getcwd()).resolve()

if script_dir.name == "notebooks":
    print(f"prject dir :: {script_dir.name} ")
    project_dir = script_dir.parent
    print(f"Project dir :: {project_dir}")
else:
    project_dir = script_dir


SOURCEDATA = project_dir / "data"
```

```python
#excel files path and import 

rewards_dir = SOURCEDATA / "Rewards"

rewards_files = list(rewards_dir.glob("*.xlsx"))
```

```python
dfs = [pd.read_excel(file) for file in rewards_files]
df = pd.concat(dfs , ignore_index=True)
```

## Data cleaning

### Dropping Columns

```python
columns_to_drop = [
     "FileName", "FileDetails", "ClearingHousePointPrice", "ClearingHouseCommissionPerTransaction",
    "TotalClearingHouse", "InvoiceNumber", "Attribute10", "Attribute9", "Attribute8", "Attribute6",
    "Attribute4", "Attribute3", "Attribute2", "Attribute1", "DiscountAmount" , "Note","MerchantName","MerchantNumber","Reference","EntityCategory",
    "PointsRedeemed","OperationType","CardTypeName","LoyaltyProgramName","TransactionType","ForeignCurrency"
]

rewards_df = df.drop(columns=columns_to_drop , errors="ignore")
```

#### Checking null values

```python
def check_nan_values(df):
    """
    Function to check null (NaN) values in each column of the DataFrame
    and print the column name along with th ecount of null (NaN) values.
    """ 
    null_counts = df.isnull().sum()
    for col , count in null_counts.items():
        print(f"{col} : {count} null (NaN) values")
```

```python
check_nan_values(rewards_df)
```

```python
#filtering Points Rewarded and Total Paid from negative values
rewards_df = rewards_df[(rewards_df['PointsRewarded'] > 0) & (rewards_df['TotalPaid'] > 0)]
```

```python
#converting TrxDate to date time
rewards_df['TrxDate'] = pd.to_datetime(rewards_df['TrxDate'])
```

## Export Cleaned to CSV

```python
#defining cleaned directory path

cleaned_dir = SOURCEDATA / "Cleaned"

#creating the directory if it does not exist
cleaned_dir.mkdir(parents=True , exist_ok=True)
```

```python
rewards_df.to_csv(cleaned_dir / "rewards_transactions_cleaned.csv", index=False)
```
